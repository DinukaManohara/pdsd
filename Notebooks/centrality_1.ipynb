```python
import networkx as nx
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('ca-HepPh.csv')
```

```python
nodes = np.concatenate((dataset['Source'].unique(), dataset['Target'].unique()))
#edges = dataset.iloc[:, :].values
```

```python
nodes = np.unique(nodes)
```

```python
for j in range(0, len(nodes)):
    dataset = dataset.replace(nodes[j], j)
    #replacer.append((int(temp_node[j]), int(j)))
    #print(replacer)
```

```python
edges = dataset.iloc[:, :].values
```

```python
nodes = np.arange(0, len(nodes))
```

```python
G = nx.Graph()
G.add_nodes_from(nodes)
G.add_edges_from(edges)
G = G.to_undirected()
```

```python
bet_cen = nx.algorithms.centrality.betweenness_centrality(G)
```

```python
G.remove_edges_from(G.selfloop_edges())
cores = nx.algorithms.core.core_number(G)

k_max = max(cores.items(), key=operator.itemgetter(1))[1]
k_min = min(cores.items(), key=operator.itemgetter(1))[1]

densities = []
for i in range(k_min, k_max + 1):
    H = nx.algorithms.core.k_core(G, k=i)
    densities.append(H.number_of_edges() / H.number_of_nodes())
    
densities
```
